# Normalization

```python
# Calling the normalization constructor

from sklearn.preprocessing import Normalizer

rescaledxdf_r_s_Arr = rescaledxdf_r_s.values

X = rescaledxdf_r_s_Arr[:,0:8]
Y = rescaledxdf_r_s_Arr[:,8]
```

```python
#Fitting and transformation
normalizer = Normalizer().fit(X)
normalizedxdf = normalizer.transform(X)
set_printoptions(precision=3)

print(normalizedxdf)
```

```python
# Change the data into a dataframe
NormalizedDF = pd.DataFrame(normalizedxdf, columns=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age'])

# Bring in the ouput feature, which is Y
NormalizedDF["Outcome"] = rescaledxdf_r_s["Outcome"]

print(NormalizedDF.head())
```
