```python
import pandas as pd
import zipfile
from io import BytesIO
```

```python
# If you use a single slash in the file path, then you need to type r before the path.
## If you use a double slash in the file path, then there is no need for r before the file path.

filepath = "C:\\Users\\MAX TECH STORE\\Documents\\DataScienceProjects\\Personal Projects\\Heart project\\Heart dataset.zip"
```

```python
# Opening the zipfile using the ZipFile function

archive = zipfile.ZipFile(filepath, 'r')
archive
```

```python
# Checking the number of files in the archive

archive.namelist()
```

```python
# Reading the heart csv

heart = archive.read('heart.csv')
heart
```

```python
# Checking the filetype of the heart dataset

type(heart)
```

```python
# Since the data is in bytes, lets convert it to pandas dataframe using the BytesIO function

data = BytesIO(heart)
df = pd.read_csv(data)
```

```python
# Checking the filetype of df

type(df)
```

```python
# Reading the top records of the dataset

df.head()
```
