```python
# Defining the pd
import pandas as pd

# Importing the file
car_sale = pd.read_csv("car-sales data.csv")
car_sale
```

```python
# Creating a column from a series
car_sale["Seats"] = pd.Series([5,5,5,5,5])
car_sale
```

```python
car_sale["Seats"].fillna(5, inplace=True)
car_sale
```

```python
# Creating a column from a list
car_sale["Fuel economy per KM"] = [3.4, 8.5, 7.2, 9.0, 5.7, 8.3, 2.9, 7.4, 5.9, 3.6]
car_sale
```

```python
# Creating column from another column
# Creating a column "Total Fuel used (L)" which will be "Odometer" divided by 100, multiplied by "Fuel per 100KM

car_sale["Total Fuel used (L)"] = car_sale["Odometer"] / 100 * car_sale["Fuel economy per KM"]
car_sale
```

```python
# Creating a column from a single value

car_sale["Wheels"] = 8
car_sale
```

```python
car_sale["Passed Road Safety"] = True
car_sale
```
